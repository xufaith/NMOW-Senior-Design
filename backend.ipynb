```python
from supabase import create_client, Client

# Replace with your Supabase project URL and anon key
supabase_url = "https://atmafjrxijoqjphoclzo.supabase.co"
supabase_key = "eyJhbGciOiJIUzI1NiIsInR5cCI6IkpXVCJ9.eyJpc3MiOiJzdXBhYmFzZSIsInJlZiI6ImF0bWFmanJ4aWpvcWpwaG9jbHpvIiwicm9sZSI6ImFub24iLCJpYXQiOjE3MzgyNjMxNDAsImV4cCI6MjA1MzgzOTE0MH0.UDfIVdt4hWaIzujfJeRZhXDgoeMe8CjnGdy6Az6aIrc"

# Initialize the Supabase client
supabase: Client = create_client(supabase_url, supabase_key)
print("Supabase client initialized")

# ----------------- CRUD Operations -----------------

# 1. Add a new ingredient
def add_new_ingredient(barcode, ingredient_name, weight, quantity, unit, expiry_date, shelf_location, category, storage_type):
    print("Adding a new ingredient...")
    response = supabase.table("ingredients").insert({
        "barcode": barcode,
        "ingredient_name": ingredient_name,
        "weight": weight,
        "quantity": quantity,
        "unit": unit,
        "expiry_date": expiry_date,
        "shelf_location": shelf_location,
        "category": category,
        "storage_type": storage_type
    }).execute()

    if response.data:
        print("Ingredient added successfully:", response.data)
    else:
        print("Error adding ingredient:", response)

# 2. Remove (delete) an existing ingredient
def remove_ingredient(barcode):
    print(f"Removing ingredient with barcode: {barcode}...")
    response = supabase.table("ingredients").delete().eq("barcode", barcode).execute()

    if response.data:
        print("Ingredient removed successfully:", response.data)
    else:
        print("Error removing ingredient:", response)

# 3. Increase ingredient stock (e.g., buy more)
def increase_stock(barcode, amount):
    print(f"Increasing stock by {amount} for ingredient with barcode: {barcode}...")
    current_quantity = supabase.table("ingredients").select("quantity").eq("barcode", barcode).execute().data[0]["quantity"]
    response = supabase.table("ingredients").update({
        "quantity": current_quantity + amount
    }).eq("barcode", barcode).execute()

    if response.data:
        print("Stock increased successfully:", response.data)
    else:
        print("Error increasing stock:", response)

# 4. Decrease ingredient stock (e.g., use some)
def decrease_stock(barcode, amount):
    print(f"Decreasing stock by {amount} for ingredient with barcode: {barcode}...")
    current_quantity = supabase.table("ingredients").select("quantity").eq("barcode", barcode).execute().data[0]["quantity"]

    if current_quantity < amount:
        print("Error: Not enough stock to decrease!")
        return

    response = supabase.table("ingredients").update({
        "quantity": current_quantity - amount
    }).eq("barcode", barcode).execute()

    if response.data:
        print("Stock decreased successfully:", response.data)
    else:
        print("Error decreasing stock:", response)

# 5. Fetch all ingredients (Read)
def fetch_all_ingredients():
    print("Fetching all ingredients...")
    response = supabase.table("ingredients").select("*").execute()

    if response.data is not None:
        print("Ingredients:", response.data)
    else:
        print("Error fetching ingredients:", response)

# ----------------- Example Calls -----------------

# Example: Adding a new ingredient
add_new_ingredient(
    barcode="xyz_123",
    ingredient_name="Milk",
    weight=2,
    quantity=10,
    unit="Liters",
    expiry_date="2025-03-01",
    shelf_location="A3",
    category="dairy",
    storage_type="fridge"
)

# Example: Removing an ingredient
remove_ingredient("xyz_123")

# Example: Increasing stock
increase_stock("qw_5728572", 5)  # Adding 5 to the ingredient with barcode 'qw_5728572'

# Example: Decreasing stock
decrease_stock("qw_5728572", 3)  # Removing 3 from the ingredient with barcode 'qw_5728572'

# Example: Fetching all ingredients
fetch_all_ingredients()
```
